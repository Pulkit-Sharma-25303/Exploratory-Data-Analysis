# **Some Mandatory Libraries to import in Data Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#**While reading files in case of error use encoding encoding_values varies according to the type and perform differently**

```python
df = pd.read_csv("/content/zomato.csv", encoding = "latin-1")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum() > 0]
```

```python
df2 = pd.read_excel("/content/Country-Code.xlsx")
df2.head()
```

```python
df3 = pd.merge(df, df2, on="Country Code", how="left")
df3.head()
```

```python
df3.Country.value_counts()
```

```python
country_name = df3.Country.value_counts().index
```

```python
country_value = df3.Country.value_counts().values
```

```python
plt.pie(country_value[:3], labels = country_name[:3], autopct="%1.2f%%")
```

**Maximum transactions on zomato are from India followed by United Kingdom and United States**

```python
ratings = df3.groupby(["Aggregate rating", "Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

**reset_index() is used for converting the numpy array into dataframe**

**size() is used to get the size of the data used**

```python
ratings
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = "Aggregate rating", y = "Rating Count",hue ="Rating color", data = ratings, palette=["grey","red","orange","yellow","green","green"])
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = "Rating color", data = ratings, palette=["grey","red","orange","yellow","green","green"])
```
